# Missing Values and Imputation

```python
import seaborn as sns
cols = music_cleaned2.columns 
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(music_cleaned2[cols].isnull(),
            cmap=sns.color_palette(colours))
```

```python
music_cleaned2[music_cleaned2.isnull().any(axis=1)].head()
```

```python
music_cleaned2.isnull().sum()
```
